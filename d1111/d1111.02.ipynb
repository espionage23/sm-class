```python
import pandas as pd
import numpy as np
df = pd.read_csv('chipotle.tsv',sep='\t')
```

```python
df.tail(2)
```

```python
df.describe()
```

```python
# 함수를 사용해서 
# item_price $부분을 없애고, float 타입으로 변경해서 입력.

def func(x):
  return float(x[1:])

# 함수호출
df['item_price'] = df['item_price'].apply(func)


df.head(3)
```

```python
# lambda  적용
df['item_price'] = df['item_price'].apply(lambda x:float(x[1:]))
df
```

```python
df.sort_values('item_price',ascending=False)
```

```python
df['item_price'].sum()
df['item_price'].mean()
df['item_price'].max()
```
