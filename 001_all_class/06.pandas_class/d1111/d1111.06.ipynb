```python
import pandas as pd
import numpy as np
df = pd.read_csv('chipotle.tsv',sep='\t')
df.tail(2)
```

```python
# 메뉴별 주문수량을 출력하시오.
df['item_name'].value_counts()
```

```python
# 메뉴별 총 주문량을 출력하시오.
len(df['order_id'].unique())
```

```python
df.info()
```

```python
# 주문당 평균 계산 금액을 출력하시오.
df['item_price'] = df['item_price'].str[1:].astype(float)
df.groupby('order_id')['item_price'].sum().mean()
```

```python
# 아이템별 가격 금액을 출력을 하시오.
filt = df['quantity'] == 1
df[filt].groupby('item_name')[['item_name','item_price']].min().sort_values('item_price')
```

```python
# 주문별, 높은 가격순으로 주문 개수를 출력하시오.
filt = df['quantity'] == 1
df[filt].groupby('item_name')[['item_name','item_price']].max().sort_values('item_price')
```
