```python
# 원그래프
values = [30,25,20,13,10,2]
explode = [0.2,0.1,0,0,0,0]
# edgecolor: 테두리 색, linewidth: 선두께
wedgeprops = {'width':0.6,'edgecolor':'black','linewidth':2}
labels = ['python','java','javascript','oracle','html','etc']
colors = ['plum','beige','skyblue','bisque','khaki','peachpuff']

# 데이터 값 : 2% 이하의 값은 출력되지 않도록 함수적용
# pct: values의 퍼센트 값 생성
# def custom_autopct(pct):
#   if pct >= 10:
#     # return ('%.1f%%' % pct)
#     return ('{:.1f}%'.format(pct))
#   else:
#     return ''
  
def custom_autopct(pct):
  return f'{pct:.1f}%' if pct >= 10 else f'{pct:.0f}'

# wedgeprops: 중간공백, pctdistance: 글자위치
plt.pie(values,labels=labels,colors=colors, startangle=90,autopct=custom_autopct,wedgeprops=wedgeprops, explode=explode, pctdistance=0.7)
plt.show()
```
