```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win 용
matplotlib.rcParams['font.size'] = 15 #글자
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('chipotle.tsv',sep='\t')
df
```

```python
# 메뉴별 주문개수
# 순서상관 없이, head(5)로 5개 막대그래프를 출력하시오.

count_df = df.groupby('item_name')['order_id'].count().head(10)

plt.figure(figsize=(30,10))
plt.bar(count_df.index,count_df)

plt.show()
```
