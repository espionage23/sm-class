```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win 용
matplotlib.rcParams['font.size'] = 15 #글자
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
### 남자만 가지고 와서 세로막대그래프 먼저 출력하시오.
df = pd.read_excel('202410연령별인구현황.xlsx',skiprows=3,usecols='B,E:Y',nrows=1,index_col='행정기관')
df2 = pd.read_excel('202410연령별인구현황.xlsx',skiprows=3,usecols='B,AB:AV',nrows=1,index_col='행정기관')
```

```python
df2.columns = df.columns
```

```python
df.loc['전국  '].str.replace(",","").astype(int)//1000
df2.loc['전국  '].str.replace(",","").astype(int)//1000

plt.figure(figsize=(20,8))
plt.title('2024년 인구현황')
plt.barh(df.columns,df.iloc[0])
plt.barh(df2.columns,-df2.iloc[0])
plt.xlabel('인구수')
plt.ylabel('연령대')
plt.legend()
plt.show()
```
