```python
import matplotlib.pyplot as plt
# 한글설장
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win 용
matplotlib.rcParams['font.size'] = 15 #글자
matplotlib.rcParams['axes.unicode_minus'] = False


# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
```

```python
import pandas as pd
import random
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# x축 : 지원번호 , y축 국어
# 선색상 : red, marker색: yellow, 테두리:blue, 선타입 :--

x = df.index
k = df['국어']
e = df['영어']
m = df['수학']
# plt.plot(x,y, color='red',marker='o' , markerfacecolor='yellow', markeredgecolor = 'blue', ls='--')

# 그래프 크기 조정 dpi: 그래프 해상도 설정 해상도가 높을수록 이미지 크기 커짐.
# facecolor: 배경색 설정
plt.figure(figsize=(20,5),dpi=200,facecolor='yellow') # x축으로 5, y축으로 20 크기 조절
# b: 선색상, v: 마커표시, :선 모양
plt.plot(x,k, label='국어점수', marker='o',ms=15, color ='red', mfc='yellow', mec='blue', ls='--')
plt.plot(x,e, label='영어점수', marker='v',ms=15)
# plt.plot(x,m, label='수학점수', marker='x',ms=15)
plt.legend()
# 그래프 파일로 저장
# plt.savefig('graph.png',dpi=100)
```
