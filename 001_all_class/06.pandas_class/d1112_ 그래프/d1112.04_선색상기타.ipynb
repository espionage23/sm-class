```python
import matplotlib.pyplot as plt
# 한글설장
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win 용
matplotlib.rcParams['font.size'] = 15 #글자
matplotlib.rcParams['axes.unicode_minus'] = False


# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
```

```python
# x,y축의 데이터
x = [1,2,3]
y = [2,4,8]
yy = [1,8,3]

# 그래프 생성
# linewidth : 선두께, color: 선색상
plt.plot(x,y, label='인원증가',color='pink',linewidth=5, marker = 'o', markersize = 10, markeredgecolor = 'red')
# marker : 지점표시 - o, x, v  markersize : 마커 크기 markerfacecolor : 마커 색상 markeredgecolor, markeredgecolor : 마커 테두리
# alpha : 투명도 설정, ls: 라인 스타일 : 점선, - 실선 --대시선 -.
# plt.plot(x,yy, label='비용증가',marker='v',markersize=10, markerfacecolor='yellow',alpha=0.4, ls='-.')

# 설정 단축표시
# b: 선색상, v: 마커표시, ; 선모양 
plt.plot(x,yy,'bv:',label='비용변동')

# 상단 타이틀
plt.title("라인그래프 Line", size=30) # size: 글자크기 변경

# 범례 : 그래프 선 설명문구, loc: 범례 위치 upper,lower,left,center,right
plt.legend(loc='upper center')
plt.legend(loc=(0.3,0.1))   # 좌표값으로 위치 설정 가능
```
