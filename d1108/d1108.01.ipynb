문자열 함수 : startswith, contains, replace, lower, upper

```python
# startswith : 해당값으로 시작되는 단어 검색
filt = df['이름'].str.startswith('김')
df[filt]

# contains : 해당값이 포함되어 있는 단어 검색
filt = df['이름'].str.contains('수')
df.loc[filt,['이름','학교','키']]

# isin() 포함이 되어 있는지 확인
langs = ['Python','Java']
file = df['SW특기'].isin(langs)
df[filt]

# lower() : 해당값을 모두 소문자로 변환
# isin() : 괄호 안에는 리스트 타입만 가능
langs = ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]

# 포함이 되어 있으면 모두 검색   
# nan 데이터 처리를 하지 않으면 에러
filt = df['SW특기'].str.contains('Java',na=False)
filt
df[filt]

# j 가 포함되어 잇는 정보 검색
filt = df['SW특기'].str.lower().str.contains('j',na=False)
df[filt]
```

null값 처리 : fillna - null값 채우기, dropna - null값 삭제하기

```python
import numpy as np

# df['학교'] = np.nan # 학교 컬럼에 NaN 데이터 입력

# fillna : 모든 컬럼의 NaN 데이터를 모두 변경, inplace=True 를 해야 적용이 됨.
df.fillna('데이터없음',inplace=True)
df
```

```python
## dropna
# axis 축방향으로 삭제 : index(row 방향), columns(columns 방향)
# how : any(1개라도 있으면 모두 row나 columns 삭제), all(모두 NaN 데이터일때 삭제)
# axis = 'index', 'columns', how='any', 'all'
df.dropna()   # index, any 가 기본 default - 해당되는 index(row)를 삭제


# df['학교'] = np.nan
df.dropna(axis='index',how='all')   # 모두 nan row가 없어서 삭제가 안됨
df.dropna(axis='columns',how='all')   # 학교 컬럼이 모두 nan이기에 삭제 됨
df.dropna(axis='index',how='any')   # 학교컬럼이 모두 nan 이기에 모든 row 삭제
df.dropna(axis='columns',how='any')   # 학교, SW특기 컬럼에 nan이 있기에 2개 컬럼 삭제

```
