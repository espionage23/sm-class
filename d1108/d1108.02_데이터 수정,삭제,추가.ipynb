```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

데이터 수정

```python
## replace : 데이터 값을 변경, inplace = True가 들어가야 저장.
# df['학교'].replace({'구로고':'가산고'},inplace=True)
# df['학교'].replace({'구로고':'가산고','디지털고':'단지고'},inplace=True)

# # 특정컬럼 값 변경
# 컬럼 = 변경값을 입력
df['SW특기'] = df['SW특기'].str.lower()   # 소문자
df['SW특기'].str.upper()    # 대문자
df['SW특기'].str.capitalize()    # 첫글자 대문자
# df['키'] = 200

# 컬럼명 = 문자 + 문자
df['학교'] = df['학교'] + '등학교'

# 특정셀 값을 변경
df.loc['1번','국어'] = 100
df.loc[['2번','3번'],['국어','영어']] = 50
```

데이터 추가

```python
# 데이터 추가
# 컬럼에 없는 컬럼 이름을 입력하면 컬럼이 추가 됨.
df['합계'] = 0
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['평균'] = df['합계']/5
df
```

조건식을 사용해서 값변경

```python
df['결과'] = 'Fail'

# 결과컬럼에 -> 조건식을 사용해서 값을 변경
filt = (df['합계']>=400)
df.loc[filt,'결과'] = 'Pass'

# 국어점수가 90점 이상이면 df['합산']+100
filt = (df['국어']>=90)
df.loc[filt,'합산'] = df.loc[filt,'합산']+100

df
```

데이터 삭제

```python
# 데이터 삭제 - drop : columns, 적용 inplace=True
# df.drop(columns=['합산'],inplace=True)

# 컬럼 2개 삭제
df.drop(columns=['국어','영어'])

# 전체 컬럼의 슬라이싱을 가지고 특정컬럼을 삭제할수 있음
df.columns[3:5]
df.drop(columns=df.columns[3:5])
```
