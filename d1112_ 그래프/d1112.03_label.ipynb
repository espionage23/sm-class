```python
import matplotlib.pyplot as plt
# 한글설장
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win 용
matplotlib.rcParams['font.size'] = 15 #글자
matplotlib.rcParams['axes.unicode_minus'] = False

# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac용
```

```python
x = [1,2,3]
y = [2,4,8]

# 그래프 생성
plt.plot(x,y)

# 상단 타이틀
plt.title("라인그래프 Line", size=30) # size: 글자크기 변경

# 그래프 축 설명
# 그래프 X축 설명 , color : 글자색 변경 , loc : 글자위치
plt.xlabel('X축 날짜', color='#8C4E37' , loc='center')  # left, center, right
plt.ylabel('Y축 날짜', color='blue', loc='bottom') # top, center, bottom

# 그래프 눈금단위 표시
plt.xticks([1,2,3])
plt.yticks([2,3,4,8,10])

plt.show()
```
