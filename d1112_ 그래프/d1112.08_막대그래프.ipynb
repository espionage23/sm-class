```python
import matplotlib.pyplot as plt
# 한글설장
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win 용
matplotlib.rcParams['font.size'] = 15 #글자
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
label = ['홍길동','유관순','강감찬']
values = [180,187,184]

# color: 색상, width: 바 크기 조절
colors = ['r','g','b']

plt.figure(figsize=(5,3))  # 그래프 크기 조절
plt.bar(label,values,color=colors, width=0.2, label='학생키') # 바 색상을 여러개로 처리
plt.xticks(label,['1번','2번','3번'],rotation=45) # X라벨 변경
plt.yticks([0,50,100,150,200,250,300])
plt.title('막대 그래프')  # 상단제목
# plt.bar(label,values,color='r') # 1개의 색상으로 처리
# plt.plot(label,values)  # 선그래프
plt.legend(loc='upper center')
plt.show()
```

막대그래프 무늬 설정

```python
colors = ['red','green','blue']
bar = plt.bar(label,values, width=0.5, color=colors,alpha=0.5)
# 막대그래프의 무늬설정
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

* 범위설정 xticks, yticks - 0부터 표시 됨.
* 특정구간부터 범위설정: xlim, ylim

```python
colors = ['red','green','blue']
plt.bar(label,values, width=0.5, color=colors,alpha=0.5)
# plt.yticks([0,100,150,200]) # 0부터 범위지정
plt.ylim(175,190) # 특정구간 범위지정
plt.show()
```

```python
label = ['홍길동','유관순','강감찬']
values = [180,187,184]
colors = ['red','green','blue']
plt.barh(label,values,color=colors,height=0.5)
plt.xlim(175,190)
plt.show()
```
