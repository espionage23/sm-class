```python
import matplotlib.pyplot as plt
# 한글설장
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # Win 용
matplotlib.rcParams['font.size'] = 15 #글자
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# x,y축의 데이터
x = [1,2,3]
y = [2,4,8]   # sm 매출액
y2 = [5,1,3]  # lg 매출액
y3 = [1,2,5]  # sk 매출액

plt.figure(figsize=(10,5))
plt.plot(x,y, label='SM 매출액', marker='o', ls='--')
plt.plot(x,y2, label='LG 매출액', marker='s', ls='-.')
plt.plot(x,y3, label='SK 매출액', marker='P', ls=':')

# 그래프 내 값출력
# for idx,txt in enumerate(y):
#   plt.text(x[idx],y[idx]+0.2,txt,ha='center')
for i in range(3):
  plt.text(x[i],y[i]+0.2,y[i],ha='center')
  plt.text(x[i],y2[i]+0.2,y2[i],ha='center')
  plt.text(x[i],y3[i]+0.2,y3[i],ha='center')

plt.title('기업별 매출액')
plt.legend(ncol=3)  # 범례출력 기본 : 세로출력
plt.show()
```
